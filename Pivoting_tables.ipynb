```python
import pandas as pd
student=pd.read_csv('student.csv')
```

```python
student.info()
```

```python
student
```

```python
pd.pivot_table(student, index=['sex','internet'], values='score')
```

```python
pd.pivot_table(student, index=['sex','internet'], values='score', aggfunc=['min','max','mean'])
```

```python
pd.pivot_table(student, index=['sex','internet'], values='score', aggfunc='mean', columns='studytime')
```

```python
pd.pivot_table(student, index=['sex',], values='score', aggfunc='mean', columns=['studytime', 'internet'], fill_value= -999)
```

```python
pd.pivot_table(student, index=['familysize'], values='score', aggfunc='mean', columns=['sex'], fill_value= -999)
```

```python
pd.pivot_table(student, index=['familysize'], values='score', aggfunc='mean', columns=['sex'], margins=True, margins_name='Average score total')
```

```python
student[student['sex']=='F'].mean(numeric_only=True)
```

```python
pd.pivot_table(student, index=['studytime'], aggfunc={'age' : ['min','max'] ,'score' : 'median'})
```

```python
pd.pivot_table(student,
              index='studytime',
              values='score',
              columns='sex',
              aggfunc=lambda range:range.max()-range.min())
```
